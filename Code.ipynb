```python
import numpy as np
import pandas as pd
```

```python
keywords=pd.read_csv('keywords.csv')
meta=pd.read_csv('movies_metadata.csv')
```

```python
def preprocess_metaid(text):
    if '-' in text:
        return 0
    else:
        return text
meta['id']=meta['id'].apply(preprocess_metaid)
meta['id']=meta['id'].astype('int64')
```

```python
final_data=final_data[['keywords','cast','crew','original_title']]
```

```python
final_data=final_data.sample(10000)
```

```python
import ast

def extracting_names(text):
    try:
        # If text is a string representation of a list of dictionaries, convert it
        if isinstance(text, str):
            text = ast.literal_eval(text)
        
        # Now extract the names if it's a valid list of dictionaries
        if isinstance(text, list):
            return [item['name'] for item in text if 'name' in item]
        else:
            return []  # Return empty if not in expected format
    except (ValueError, SyntaxError):
        return []

final_data['keywords']=final_data['keywords'].apply(extracting_names)
```

```python
import ast

def extract_cast_names(text):
    try:
        # Convert the string to a list of dictionaries if needed
        if isinstance(text, str):
            text = ast.literal_eval(text)
        
        # Extract the 'name' field from each dictionary in the list
        if isinstance(text, list):
            return [item['name'] for item in text if 'name' in item]
        else:
            return []  # Return empty if not in the expected format
    except (ValueError, SyntaxError):
        # Handle cases where conversion fails
        return []

final_data['cast']=final_data['cast'].apply(extract_cast_names)
```

```python
final_data['crew']
```

```python
def extract_crew_names(text):
    try:
        # Convert the string to a list of dictionaries if needed
        if isinstance(text, str):
            text = ast.literal_eval(text)
        
        # Extract the 'name' field from each dictionary in the list
        if isinstance(text, list):
            return [item['name'] for item in text if 'name' in item]
        else:
            return []  # Return empty if not in the expected format
    except (ValueError, SyntaxError):
        # Handle cases where conversion fails
        return []

final_data['crew']=final_data['crew'].apply(extract_crew_names)
```

```python
final_data['Content']=final_data['keywords']+final_data['cast']+final_data['crew']
```

```python
final_data.drop(['keywords','cast','crew'],axis=1,inplace=True)
```

```python
final_data['Content']=final_data['Content'].apply(lambda x: ' '.join(x))
```

```python
import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
ps=PorterStemmer()
final_data['Content']=final_data['Content'].apply(lambda x:''.join([ps.stem(i) for  i in  x if i not in stopwords.words('english') ]))
```

```python
final_data
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf=TfidfVectorizer()
vectors=tf.fit_transform(final_data['Content']).toarray()
```

```python
vectors[0]
```

```python
vectors.shape
```

```python
pivot_table = pd.DataFrame(vectors, index=final_data['original_title'], columns=tf.get_feature_names_out())
```

```python
pivot_table
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(pivot_table)
```

```python
def recommend_movies(movie_name):
    """Recommends similar movies based on a given movie name.

    Args:
        movie_name (str): The name of the movie to find similar recommendations for.

    Returns:
        None
    """

    # Find the index of the movie in the dataset
    index = np.where(pivot_table.index == movie_name)[0][0]

    # Find the most similar movies to the given movies
    similar_movies = sorted(list(enumerate(similarity[index])), key=lambda x: x[1], reverse=True)[1:6]

    # Print the recommended movies
    for movie_index, _ in similar_movies:  # unpack the tuple (index, similarity)
        print(pivot_table.index[movie_index])
```

```python
recommend_movies('Single White Female 2: The Psycho')
```
